# PRUEBA

````python
%%manim SquareToCircle -qm -v WARNING

class SquareToCircle(Scene):
    def construct(self):
        circle = Circle()
        square = Square()
        square.flip(RIGHT)
        square.rotate(-3 * TAU / 8)
        circle.set_fill(PINK, opacity=0.5)

        self.play(ShowCreation(square))
        self.play(Transform(square, circle))
        self.play(FadeOut(square))
````

````python
%%manim HolaMundo -qm -v WARNING

class HolaMundo(Scene):
    def construct(self):

        texto = Tex("¡Hola Mundo!")

        # Tex y Text funciona mejor quue TextMobject
        # Tex pone el texto con tipografía de LaTeX y Text con tipografía normal

        self.play(Write(texto))
        self.wait(3)
````

````python
%%manim Bienvenida_Ciencias -qm -v WARNING

class Bienvenida_Ciencias(Scene):
    def construct(self):

        bienvenida = MathTex("\\textup{Bienvenidxs a } \\mathbb{C} 1 \\varepsilon (\\eta) \\mathbb{C} 1 \\alpha \\int_{x}" )
        bienvenida.scale(2)

        # MathTex funciona mejor que TexMobject

        self.play(Write(bienvenida))
        self.wait(3)
````

````python
%%manim Demostracion1 -ql -v WARNING

# ql calidad baja
# qh calidad alta

class Demostracion1(Scene):
    def construct(self):

        serie = MathTex("\\sum_{n=1}^{\\infty} \\frac{1}{4^{n}} =  \\frac{1}{4^{1}} + \\frac{1}{4^{2}} + \\frac{1}{4^{3}}  + \\frac{1}{4^{4}} + \\frac{1}{4^{5}} + \\frac{1}{4^{6}} + \\cdots = \\frac{1}{3}" )

        demostracion = Tex("Demostración")
        demostracion.scale(2)

        # PLANO CARTESIANO
        # plano = NumberPlane().add_coordinates()
        # self.add(plano)

        # PRIMER CUADRADO PRINCIPAL
        cuadrado1 = Square(side_length = 4)
        cuadrado1.move_to([-5, 0, 0])

        # UN CUARTO AZUL
        cuartoAzul = Square(side_length = 2)
        cuartoAzul.set_fill(BLUE, opacity = 0.5)
        cuartoAzul.set_stroke(BLUE, width = 3)
        cuartoAzul.move_to([-6, -1, 0])

        # UN 16 AZUL
        diezseisAzul = Square(side_length = 1)
        diezseisAzul.set_fill(BLUE, opacity = 0.5)
        diezseisAzul.set_stroke(BLUE, width = 3)
        diezseisAzul.move_to([-4.5, -1.5, 0])

        # UN 32 AZUL
        treintadosAzul = Square(side_length = 0.5)
        treintadosAzul.set_fill(BLUE, opacity = 0.5)
        treintadosAzul.set_stroke(BLUE, width = 3)
        treintadosAzul.move_to([-3.75, -1.75, 0])

        #  UN 64 AZUL
        seiscuatroAzul = Square(side_length = 0.25)
        seiscuatroAzul.set_fill(BLUE, opacity = 0.5)
        seiscuatroAzul.set_stroke(BLUE, width = 3)
        seiscuatroAzul.move_to([-3.375, -1.875, 0])

        #  UN 128 AZUL
        cienochoAzul = Square(side_length = 0.125)
        cienochoAzul.set_fill(BLUE, opacity = 0.5)
        cienochoAzul.set_stroke(BLUE, width = 3)
        cienochoAzul.move_to([-3.1875, -1.9375, 0])  

        #  UN 256 AZUL
        doscincuentaAzul = Square(side_length = 0.0625)
        doscincuentaAzul.set_fill(BLUE, opacity = 0.5)
        doscincuentaAzul.set_stroke(BLUE, width = 3)
        doscincuentaAzul.move_to([-3.09375, -1.96875, 0])  

        # SEGUNDO CUADRADO PRINCIPAL
        cuadrado2 = Square(side_length = 4)
        cuadrado2.move_to([0, 0, 0])

        # UN CUARTO ROJO
        cuartoRojo =  Square(side_length = 2)
        cuartoRojo.set_fill(RED, opacity = 0.5)
        cuartoRojo.set_stroke(RED, width = 3)
        cuartoRojo.move_to([1,1,0])

        # UN 16 ROJO
        diezseisRojo = Square(side_length = 1)
        diezseisRojo.set_fill(RED, opacity = 0.5)
        diezseisRojo.set_stroke(RED, width = 3)
        diezseisRojo.move_to([1.5, -0.5, 0])

        # UN 32 ROJO
        treintadosRojo = Square(side_length = 0.5)
        treintadosRojo.set_fill(RED, opacity = 0.5)
        treintadosRojo.set_stroke(RED, width = 3)
        treintadosRojo.move_to([1.75, -1.25, 0])

        # UN 64 ROJO
        seiscuatroRojo = Square(side_length = 0.25)
        seiscuatroRojo.set_fill(RED, opacity = 0.5)
        seiscuatroRojo.set_stroke(RED, width = 3)
        seiscuatroRojo.move_to([1.875, -1.625, 0])

        #  UN 128 ROJO
        cienochoRojo = Square(side_length = 0.125)
        cienochoRojo.set_fill(RED, opacity = 0.5)
        cienochoRojo.set_stroke(RED, width = 3)
        cienochoRojo.move_to([1.9375, -1.8125, 0])  

        #  UN 256 ROJO
        doscincuentaRojo = Square(side_length = 0.0625)
        doscincuentaRojo.set_fill(RED, opacity = 0.5)
        doscincuentaRojo.set_stroke(RED, width = 3)
        doscincuentaRojo.move_to([1.96875, -1.90625, 0])  

        # TERCER CUADRADO PRINCIPAL
        cuadrado3 = Square(side_length = 4)
        cuadrado3.move_to([5, 0, 0])

        # UN CUARTO VERDE
        cuartoVerde =  Square(side_length = 2)
        cuartoVerde.set_fill(GREEN, opacity = 0.5)
        cuartoVerde.set_stroke(GREEN, width = 3)
        cuartoVerde.move_to([4,1,0])

        # UN 16 VERDE
        diezseisVerde = Square(side_length = 1)
        diezseisVerde.set_fill(GREEN, opacity = 0.5)
        diezseisVerde.set_stroke(GREEN, width = 3)
        diezseisVerde.move_to([5.5, -0.5, 0])

        # UN 32 VERDE
        treintadosVerde = Square(side_length = 0.5)
        treintadosVerde.set_fill(GREEN, opacity = 0.5)
        treintadosVerde.set_stroke(GREEN, width = 3)
        treintadosVerde.move_to([6.25, -1.25, 0])

        # UN 64 VERDE
        seiscuatroVerde = Square(side_length = 0.25)
        seiscuatroVerde.set_fill(GREEN, opacity = 0.5)
        seiscuatroVerde.set_stroke(GREEN, width = 3)
        seiscuatroVerde.move_to([6.625, -1.625, 0])

        #  UN 128 VERDE
        cienochoVerde = Square(side_length = 0.125)
        cienochoVerde.set_fill(GREEN, opacity = 0.5)
        cienochoVerde.set_stroke(GREEN, width = 3)
        cienochoVerde.move_to([6.8125, -1.8125, 0])  

        #  UN 256 VERDE
        doscincuentaVerde = Square(side_length = 0.0625)
        doscincuentaVerde.set_fill(GREEN, opacity = 0.5)
        doscincuentaVerde.set_stroke(GREEN, width = 3)
        doscincuentaVerde.move_to([6.90625, -1.90625, 0])  

        # ECUACIONES
        primeraEcuacion = MathTex("S = \\frac{1}{4^{1}}" )
        primeraEcuacion.move_to([0, -3, 0])

        segundaEcuacion = MathTex("S = \\frac{1}{4^{1}} + \\frac{1}{4^{2}} " )
        segundaEcuacion.move_to([0, -3, 0])

        terceraEcuacion = MathTex("S = \\frac{1}{4^{1}} + \\frac{1}{4^{2}} + \\frac{1}{4^{3}} " )
        terceraEcuacion.move_to([0, -3, 0])

        cuartaEcuacion = MathTex("S = \\frac{1}{4^{1}} + \\frac{1}{4^{2}} + \\frac{1}{4^{3}}  + \\frac{1}{4^{4}} " )
        cuartaEcuacion.move_to([0, -3, 0])

        quintaEcuacion = MathTex("S = \\frac{1}{4^{1}} + \\frac{1}{4^{2}} + \\frac{1}{4^{3}}  + \\frac{1}{4^{4}} + \\frac{1}{4^{5}}" )
        quintaEcuacion.move_to([0, -3, 0])

        sextaEcuacion = MathTex("S = \\frac{1}{4^{1}} + \\frac{1}{4^{2}} + \\frac{1}{4^{3}}  + \\frac{1}{4^{4}} + \\frac{1}{4^{5}} + \\frac{1}{4^{6}}" )
        sextaEcuacion.move_to([0, -3, 0])

        septimaEcuacion = MathTex("S = \\frac{1}{4^{1}} + \\frac{1}{4^{2}} + \\frac{1}{4^{3}}  + \\frac{1}{4^{4}} + \\frac{1}{4^{5}} + \\frac{1}{4^{6}} + \\cdots + \\frac{1}{4^{n}}" )
        septimaEcuacion.move_to([0, -3, 0])

        ### INICIO ###

        self.play(Write(serie))
        self.wait(3)
        self.play(FadeOut(serie))

        self.play(Write(demostracion))
        self.wait(3)
        self.play(FadeOut(demostracion))

        ###  PRIMERA ANIMACION ###

        # Cuadrados Principales
        self.play(ShowCreation(cuadrado1), ShowCreation(cuadrado2), ShowCreation(cuadrado3), run_time = 1)

        # Cuartos
        self.play(ShowCreation(cuartoAzul), ShowCreation(cuartoRojo), ShowCreation(cuartoVerde), run_time = 1)

        self.play(Write(primeraEcuacion))

        # DIES Y SEIS
        self.play(ShowCreation(diezseisAzul), ShowCreation(diezseisRojo), ShowCreation(diezseisVerde), run_time = 1)

        self.play(Transform(primeraEcuacion, segundaEcuacion))

        # TREINTA Y DOS
        self.play(ShowCreation(treintadosAzul), ShowCreation(treintadosRojo), ShowCreation(treintadosVerde), run_time = 1)

        self.play(Transform(primeraEcuacion, terceraEcuacion))

        # SESENTA Y CUATRO
        self.play(ShowCreation(seiscuatroAzul), ShowCreation(seiscuatroRojo),ShowCreation(seiscuatroVerde),  run_time = 1)

        self.play(Transform(primeraEcuacion, cuartaEcuacion))

        # 128
        self.play(ShowCreation(cienochoAzul), ShowCreation(cienochoRojo), ShowCreation(cienochoVerde), run_time = 1)

        self.play(Transform(primeraEcuacion, quintaEcuacion))

        # 256
        self.play(ShowCreation(doscincuentaAzul),  ShowCreation(doscincuentaRojo), ShowCreation(doscincuentaVerde), run_time = 1)
        self.play(Transform(primeraEcuacion, sextaEcuacion))

        self.wait(2)

        self.play(Transform(primeraEcuacion, septimaEcuacion))

        self.wait(4)

        ### JUNTAR CUADRADOS


        # PRIMER CUADRADO PRINCIPAL
        cuadrado11 = Square(side_length = 4)

        # UN CUARTO AZUL
        cuartoAzul1 = Square(side_length = 2)
        cuartoAzul1.set_fill(BLUE, opacity = 0.5)
        cuartoAzul1.set_stroke(BLUE, width = 3)

        # UN 16 AZUL
        diezseisAzul1 = Square(side_length = 1)
        diezseisAzul1.set_fill(BLUE, opacity = 0.5)
        diezseisAzul1.set_stroke(BLUE, width = 3)

        # UN 32 AZUL1
        treintadosAzul1 = Square(side_length = 0.5)
        treintadosAzul1.set_fill(BLUE, opacity = 0.5)
        treintadosAzul1.set_stroke(BLUE, width = 3)

        #  UN 64 AZUL1
        seiscuatroAzul1 = Square(side_length = 0.25)
        seiscuatroAzul1.set_fill(BLUE, opacity = 0.5)
        seiscuatroAzul1.set_stroke(BLUE, width = 3)

        #  UN 128 AZUL1
        cienochoAzul1 = Square(side_length = 0.125)
        cienochoAzul1.set_fill(BLUE, opacity = 0.5)
        cienochoAzul1.set_stroke(BLUE, width = 3)

        #  UN 256 AZUL1
        doscincuentaAzul1 = Square(side_length = 0.0625)
        doscincuentaAzul1.set_fill(BLUE, opacity = 0.5)
        doscincuentaAzul1.set_stroke(BLUE, width = 3)

        # TERCER CUADRADO PRINCIPAL
        cuadrado31 = Square(side_length = 4)

        # UN CUARTO VERDE
        cuartoVerde1 =  Square(side_length = 2)
        cuartoVerde1.set_fill(GREEN, opacity = 0.5)
        cuartoVerde1.set_stroke(GREEN, width = 3)

        # UN 16 VERDE1
        diezseisVerde1 = Square(side_length = 1)
        diezseisVerde1.set_fill(GREEN, opacity = 0.5)
        diezseisVerde1.set_stroke(GREEN, width = 3)

        # UN 32 VERDE1
        treintadosVerde1 = Square(side_length = 0.5)
        treintadosVerde1.set_fill(GREEN, opacity = 0.5)
        treintadosVerde1.set_stroke(GREEN, width = 3)

        # UN 64 VERDE1
        seiscuatroVerde1 = Square(side_length = 0.25)
        seiscuatroVerde1.set_fill(GREEN, opacity = 0.5)
        seiscuatroVerde1.set_stroke(GREEN, width = 3)

        #  UN 128 VERDE1
        cienochoVerde1 = Square(side_length = 0.125)
        cienochoVerde1.set_fill(GREEN, opacity = 0.5)
        cienochoVerde1.set_stroke(GREEN, width = 3)

        #  UN 256 VERDE1
        doscincuentaVerde1 = Square(side_length = 0.0625)
        doscincuentaVerde1.set_fill(GREEN, opacity = 0.5)
        doscincuentaVerde1.set_stroke(GREEN, width = 3)


        #  AZULES

        cuadrado11.move_to([0, 0, 0])
        cuartoAzul1.move_to([-1, -1, 0])
        diezseisAzul1.move_to([0.5, -1.5, 0])
        treintadosAzul1.move_to([1.25, -1.75, 0])
        seiscuatroAzul1.move_to([1.625, -1.875, 0])
        cienochoAzul1.move_to([1.8125, -1.9375, 0]) 
        doscincuentaAzul1.move_to([1.90625, -1.96875, 0])

        # VERDES
        cuadrado31.move_to([0, 0, 0])
        cuartoVerde1.move_to([-1, 1, 0])
        diezseisVerde1.move_to([0.5, -0.5, 0])
        treintadosVerde1.move_to([1.25, -1.25, 0])
        seiscuatroVerde1.move_to([1.625, -1.625, 0])
        cienochoVerde1.move_to([1.8125, -1.8125, 0]) 
        doscincuentaVerde1.move_to([1.90625, -1.90625, 0])

        # ECUACIONES FINALES

        sumaTotal = MathTex("3S = 1")
        sumaTotal.move_to([0, -3, 0])

        sumaFinal = MathTex("S = \\frac{1}{3}")
        sumaFinal.move_to([0, -3, 0])

        self.play(Transform(cuadrado1, cuadrado11),  Transform(cuartoAzul, cuartoAzul1), Transform(diezseisAzul, diezseisAzul1), Transform(treintadosAzul, treintadosAzul1), Transform(seiscuatroAzul, seiscuatroAzul1), Transform(cienochoAzul, cienochoAzul1), Transform(doscincuentaAzul, doscincuentaAzul1), run_time = 1)
        self.play(Transform(cuadrado3, cuadrado31),  Transform(cuartoVerde, cuartoVerde1), Transform(diezseisVerde, diezseisVerde1), Transform(treintadosVerde, treintadosVerde1), Transform(seiscuatroVerde, seiscuatroVerde1), Transform(cienochoVerde, cienochoVerde1), Transform(doscincuentaVerde, doscincuentaVerde1), run_time = 1)
        
        self.play(Transform(primeraEcuacion, sumaTotal))
        self.wait(2)
        self.play(Transform(primeraEcuacion, sumaFinal))

        self.wait(4)
````
